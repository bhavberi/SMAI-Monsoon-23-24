```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
```

```python
dataset_path = "../Datasets/SMAI-Dataset-problem-4.3/colors.pkl"
k=7
random_init_state = 13
```

```python
with open(dataset_path, 'rb') as f:
    data = pickle.load(f)
```

```python
plt.scatter(data[:, 0], data[:, 1])
```

```python
gm = GaussianMixture(n_components=k, random_state=random_init_state, init_params='k-means++')
```

```python
labels = gm.fit_predict(data)
```

```python
plt.scatter(data[:, 0], data[:, 1], c=labels)
```

```python
def generate_sample_dataset(n, means, covariances, num_samples_per_component=None) -> np.ndarray:
    if num_samples_per_component is None:
        num_samples_per_component = n
    
    sample_dataset = []
    
    for i in range(n):
        mean = means[i]
        covariance = covariances[i]
        
        # Sample data points from the current component
        component_samples = np.random.multivariate_normal(mean, covariance, num_samples_per_component)
        
        sample_dataset.extend(component_samples)
    
    return np.array(sample_dataset)
```

```python
new_data = generate_sample_dataset(7, gm.means_, gm.covariances_)
```

```python
plt.scatter(new_data[:, 0], new_data[:, 1], c=gm.predict(new_data))
```

> #### Oberservation

As we can see the newly generated data, is equivalent to the original given data, in terms of representation (of classes). It covers up all of the gaussian data, and is a good representation of the original data.
